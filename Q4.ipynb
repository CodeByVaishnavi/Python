```python
from datetime import datetime

# Function to validate email
def is_valid_email(email):
    if "@" in email and "." in email:
        return True
    return False

# Function to validate phone number
def is_valid_phone(phone):
    if phone.isdigit() and len(phone) == 10:
        return True
    return False

# Function to calculate age based on date of birth
def calculate_age(dob):
    dob_date = datetime.strptime(dob, "%Y-%m-%d")
    today = datetime.today()
    age = today.year - dob_date.year
    if (today.month, today.day) < (dob_date.month, dob_date.day):
        age -= 1
    return age

# Start taking user inputs
print("Welcome to Fitness Membership Registration!")

name = input("Enter your Name: ")

# Validate age input
while True:
    age = input("Enter your Age: ")
    if age.isdigit() and 1 <= int(age) <= 100:
        age = int(age)
        break
    else:
        print("Please enter a valid age between 1 and 100.")

gender = input("Enter your Gender (Male/Female/Other): ")

# Validate email input
while True:
    email = input("Enter your Email: ")
    if is_valid_email(email):
        break
    else:
        print("Please enter a valid email address.")

password = input("Enter your Password: ")

# Validate phone number input
while True:
    phone = input("Enter your Phone Number (10 digits): ")
    if is_valid_phone(phone):
        break
    else:
        print("Please enter a valid 10-digit phone number.")

dob = input("Enter your Date of Birth (YYYY-MM-DD): ")

# Validate preferred fitness goal
while True:
    preferred_goal = input("Enter Preferred Fitness Goal (weightloss/muscle gain): ").lower()
    if preferred_goal in ["weightloss", "muscle gain"]:
        break
    else:
        print("Please enter 'weightloss' or 'muscle gain'.")

# Validate workout timing
while True:
    preferred_timing = input("Enter Preferred Workout Timing (am/pm): ").lower()
    if preferred_timing in ["am", "pm"]:
        break
    else:
        print("Please enter 'am' or 'pm'.")

# Get weight and height
weight = float(input("Enter your Weight (in kg): "))
height = float(input("Enter your Height (in cm): "))

# Validate membership plan
while True:
    membership_plan = input("Enter Membership Plan (monthly/quarterly/yearly): ").lower()
    if membership_plan in ["monthly", "quarterly", "yearly"]:
        break
    else:
        print("Please enter 'monthly', 'quarterly', or 'yearly'.")

# Validate if already a user
while True:
    already_user = input("Are you already a user? (yes/no): ").lower()
    if already_user in ["yes", "no"]:
        break
    else:
        print("Please enter 'yes' or 'no'.")

# Calculate age from DOB
calculated_age = calculate_age(dob)

# Store data in a dictionary
user_data = {
    "Name": name,
    "Age": age,
    "Gender": gender,
    "Email": email,
    "Password": password,
    "Phone": phone,
    "DOB": dob,
    "Calculated Age": calculated_age,
    "Preferred Goal": preferred_goal,
    "Preferred Timing": preferred_timing,
    "Weight": weight,
    "Height": height,
    "Membership Plan": membership_plan,
    "Already a User": already_user,
}

print("\nUser Data Stored Successfully!")
print(user_data)
```
